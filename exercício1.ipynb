```python
# Calculadora em Python


print("\n******************* Python Calculator *******************")

def adição(x,y):
  return x + y

def subt(x,y):
  return x - y

def mult(x,y):
  return x * y

def div(x,y):
  return x / y


print('Selecione a opção desejada:')
print('1 para Adição')
print('2 para subtração')
print('3 para multiplicação')
print('4 para divisão')

escolha = int(input('Qual a sua escolha? [1,2,3,4] '))

n1 = int(input('Digite o primeiro número: '))
n2 = int(input('Digite o segundo número: '))

print('-'*50)




if escolha == 1:
  print(f'{n1} + {n2} = {adição(n1,n2)}')

elif escolha == 2:
  print(f'{n1} - {n2} = {subt(n1,n2)}')

elif escolha == 3:
  print(f'{n1} x {n2} = {mult(n1,n2)}')

elif escolha == 4:
  print(f'{n1} / {n2} = {div(n1,n2)}')

else:
  print('Opção inválida!')
```
