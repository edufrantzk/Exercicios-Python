```python
#Descubra a tabuada de vários números, o programa é interrompido quando o núero solicitado for negativo
n= 0
c= 0
while True:
  n = int(input('Digite um número para saber sua tabuada: '))
  if n < 0:
    break
  for c in range(10):
    c = c + 1
    r = n * c
    print(f'{c} x {n} = {r}')
print('Tabuada encerrada, volte sempre!')
```
