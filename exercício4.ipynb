```python
from random import randint
v = 0
print('Vamos jogar par ou impar!')
while True:
  pc = randint(1, 10)
  n = int(input('Digite um número: '))
  resultado = pc + n
  pi = str(input('PAR OU ÍMPAR? ')).strip().upper()[0]
  print(f'Você escolheu {n} e o computador escolheu {pc} o total deu {resultado}')
  print(f'Deu par' if resultado % 2 == 0 else 'Deu ímpar')
  if pi in 'pP':
    if resultado % 2 == 0:
      print('Você ganhou!')
      v = v + 1 #para somar as vitórias
    else:
      print('Você perdeu!') 
      break
  elif pi in 'Ii':
    if resultado % 2 == 1:
      print('Você ganhou!')
      v = v + 1 #para somar as vitórias
    else:
      print('Você perdeu!')
      break
  print('Vamos tentar novamente!')
print(f'Game over! Você venceu {v} vezes.')
```
