```python
#Jogando jokenpô com o computador

import random

você = str(input('O que você vai escolher? pedra, papel ou tesoura?'))
#criando a escolha aleatória do computador para disputar com você
computador = random.choice(['tesoura','pedra','papel'])
print(f'O computador escolheu:{computador}.')

#Condição para o computador ganhar:
if computador == 'tesoura' and você == 'papel' or computador == 'pedra' and você == 'tesoura' or computador == 'papel' and você == 'pedra':
  print('O computador ganhou e você perdeu!')
#Condição para você ganhar:
elif você == 'tesoura' and computador == 'papel' or você == 'pedra' and computador == 'tesoura' or você == 'papel' and computador == 'pedra':
  print('Parabéns, Você ganhou do computador!')
#Em caso de empate:
else:
  print('Empatou! Você e o computador fizeram a mesma escolha!')
```
